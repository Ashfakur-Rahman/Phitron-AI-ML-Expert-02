# Problem 2

```python
def batch_generator(lst, batch_size):
    for i in range(0, len(lst), batch_size):
        yield lst[i:i + batch_size]
x = batch_generator([1,2,3,4,5,6,7,8], 3)
print(next(x))
print(next(x))
print(next(x))
#print(next(x))
```
